```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(X_train ,y_train) ,(X_test,y_test) =keras.datasets.mnist.load_data()
```

## Do Normalization Scale on the Array of X_test and X_train ( from 0 : 1)

```python
X_train_norm=X_train/255
X_test_norm =X_test/255
```

```python
X_train_norm[0] # numbers of the Array between ( 0 : 1 )
```

digits_nn.jpg

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten (input_shape=(28, 28),name='input_Layer'),
    tf.keras.layers.Dense(120,activation='relu',name='hidden_Layer'), # nuerons (120)
   tf.keras.layers.Dense(10,activation='sigmoid',name='output_Layer') # from 0 to 9
])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train_norm, y_train, epochs=10)
model.summary()
```

```python
model.evaluate(X_test_norm,y_test)
```

```python
y_predict=model.predict(X_test_norm)
```

```python
y_predict[0]
```

```python
np.argmax(y_predict[0])
```

```python
X_test[0]
```

```python
y_predict_labels=[np.argmax(i) for i in y_predict]
```

```python
y_predict_labels [0 :5]
```

```python
for i in X_test[0:5]:
  plt.matshow(i)
  plt.show()
```

```python
cm =tf.math.confusion_matrix(labels=y_test, predictions=y_predict_labels)
cm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True ,fmt='d' )
plt.xlabel('predictions')
plt.ylabel('truth')
plt.show()
```
